```python
def filtros(x):
    '''Filtra as palavras desejadas, removendo tudo que vem antes ou depois
       arg: x que serão as strings contidas na coluna do DataFrame
       return: Irá retornar somente o padrão desejado sem as informações a parte, caso não contenha o padrão irá retornar nulo
       ex: [if 'palavra' in x.lower():
        return  re.sub('.*palavra.*', 'palavra',x, flags = re.I)]'''
    if 'surfing' in x.lower():
        return  re.sub('.*surfing.*', 'surfing',x, flags = re.I)
    if 'swimming' in x.lower():
        return re.sub('.*swimming.*', 'swimming',x, flags = re.I)
    if 'fishing' in x.lower():
         return re.sub('.*fishing.*', 'fishing',x, flags = re.I)
    if 'diving' in x.lower():
        return re.sub('.*diving.*', 'diving',x, flags = re.I)
    if 'spearfishing' in x.lower():
        return re.sub('.*spearfishing.*', 'spearfishing',x, flags = re.I)
    if 'wading' in x.lower():
        return re.sub('.*wading.*', 'wading',x, flags = re.I)
    if 'bathing' in x.lower():
        return re.sub('.*wading.*', 'wading',x, flags = re.I)
    if 'standing' in x.lower():
        return re.sub('.*standing.*', 'standing',x, flags = re.I)
    if 'snorkeling' in x.lower():
        return re.sub('.*snorkeling.*', 'snorkeling',x, flags = re.I)
    if 'boarding' in x.lower():
        return re.sub('.*boarding.*', 'boarding',x, flags = re.I)
    if 'kayaking' in x.lower():
        return re.sub('.*kayaking.*', 'kayaking',x, flags = re.I)
    if 'treading water ' in x.lower():
        return re.sub('.*treading water.*','treading water',x, flags = re.I)
    if 'walking' in x.lower():
        return re.sub('.*walking.*','walking',x, flags = re.I)
    else:
        return 
```

```python
data['Activity'] = data["Activity"].astype(str).apply(filtros)#transforma a coluna em string e aplica os filtros da função
data.dropna(subset = ['Activity'], inplace = True) #remove as linhas nulas da coluna
```

```python
data['Date'] = data["Date"].astype(str).apply(lambda x: re.findall('\d{2}-?\w{2,3}-?\d{4}',x)[0] if len(re.findall('\d{2}-?\w{2,3}-?\d{4}',x)) > 0 else np.nan)
#define um pdrão de busca para as datas
data.dropna(subset = ["Date"],inplace = True )# remove os nulos da coluna 
```
